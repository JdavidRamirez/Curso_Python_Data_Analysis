 # 1. Manejo de valores  nulos

 ## Identificación de valores nulos

### **Cuándo ELIMINAR los datos nulos**
Eliminar datos es riesgoso porque pierdes información. Solo deberías hacerlo si:
*  La cantidad de valores nulos es muy alta (ejemplo: más del 60% de una columna).
*  La columna no es importante para el análisis.
*  El registro no tiene sentido sin ese valor.

Podemos detectar valores nulos en un DataFrame utilizando `isnull()` o `isna().`

```python
import pandas as pd

# Cargar dataset
orders = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQkbxvR3vVEv8BKp7W1ZiuyzU_Wjow8iAm-Uu_VgjEfqzLUqU3-eGNbVgvj5Xtyih7FNzJJ_GXLLNuO/pub?gid=396022667&single=true&output=csv")


orders.info()
```

```python
# Identificar valores nulos
orders.isnull().sum()
```

```python
# eliminar las columnas que tienen todos los valores NaN
orders = orders.dropna(how='all', axis=1)
orders.info()
```

### 2. Imputación con un valor específico `fillna()`

```python
#1. Rellenar valores nulos con un número específico

data = {'A': [1, None, 3], 'B': [None, 5, 6]}
df = pd.DataFrame(data)
df
```

```python
#Rellenar valores nulos con un número específico
df_filled = df.fillna(0)
df_filled
```

```python
# Rellena NaN con el promedio de cada columna
df['B']=df['B'].fillna(df['B'].mean())
df
```

### 3. Usar forward fill (ffill) para rellenar con el valor anterior



```python
data = {'A': [1, None, 3, None, 5,7], 'B': [None, 5, 6,None,None,12]}
df = pd.DataFrame(data)


df
```

```python
df_filled = df.fillna(method="ffill")
df_filled
```

```python
#Usar backward fill (bfill) para rellenar con el siguiente valor
df_filled = df.fillna(method="bfill")
```

```python
df_filled
```

### 4. Rellenar con un valor de texto fijo

```python
data = {'Nombre': ['Alice', None, 'Charlie', 'David'],
        'Ciudad': ['NY', 'LA', None, 'SF']}

df = pd.DataFrame(data)
df
# Rellenar valores nulos con 'Desconocido'
df_filled = df.fillna("Desconocido")

df_filled
```

 # 3. Combinación de dataframes

 ##  **Concatenación** : `pd.concat()`



```python
"""
Ejemplo 1: Concatenación Vertical (por Filas)
Supongamos que tenemos dos DataFrame con datos de temperatura en diferentes meses:
"""

import pandas as pd

df1 = pd.DataFrame({
    'Mes': ['Enero', 'Febrero'],
    'Temperatura': [30, 32]
})

df2 = pd.DataFrame({
    'Mes': ['Marzo', 'Abril'],
    'Temperatura': [35, 38]
})

df_concatenado = pd.concat([df1, df2], ignore_index=True)
print(df_concatenado)


# ignore_index=True reinicia los indices para evitar duplicados
```

```python
"""
Ejemplo 2: Concatenación Horizontal (por Columnas)
"""

df3 = pd.DataFrame({
    'Humedad': [80, 85, 78, 90]
})

df_concatenado_horizontal = pd.concat([df_concatenado, df3], axis=1)

# Nota: axis=1 concatena las columnas en lugar de las filas.
df_concatenado_horizontal

```

 ## **Unión con** `merge()`


```python
"""
La función merge() es similar a un JOIN en SQL. Permite combinar DataFrame en base a una clave común.

Ejemplo 3: merge() con inner join
"""

df4 = pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Precipitacion': [120, 95, 110, 105]
})

df_merged = pd.merge(df_concatenado, df4, on='Mes', how='inner')

#Aquí usamos on='Mes' para unir ambos DataFrame usando la columna "Mes".
df_merged
```

## `left join` con valores faltantes.

```python
df5 = pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Mayo'],
    'Viento': [10, 12, 15, 8]
})

df_left = pd.merge(df_concatenado, df5, on='Mes', how='left')

#Nota: Como "Abril" no está en df5, su valor en la columna "Viento" aparece como NaN.

df_left
```

## Unión con `join()`

`join()` se usa cuando queremos fusionar DataFrame usando el índice.

```python
df6 = pd.DataFrame({
    'Viento': [10, 12, 15, 8]
}, index=['Enero', 'Febrero', 'Marzo', 'Abril'])

df_indexed = df_concatenado.set_index('Mes')
df_joined = df_indexed.join(df6)

#join() asocia los datos usando los índices en lugar de una columna específica
df_joined
```
