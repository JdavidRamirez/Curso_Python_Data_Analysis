 # 2. Barras, histogramas, lineas y dispersión


## Gráfico de Barras

Un gráfico de barras es una representación gráfica de datos numéricos que se utiliza para comparar cantidades o frecuencias. Se compone de barras verticales u horizontales, donde el tamaño de cada barra es proporcional al valor que representa.

### ¿Para qué sirve ?

* Permite comparar visualmente cantidades o frecuencias.
* Facilita la interpretación de los datos.
* Ayuda a analizar y tomar decisiones basadas en los datos.
* Se utiliza en diversas disciplinas, como ciencias sociales, economía, administración de empresas, ciencias de la salud, entre otras.

```python
co2_by_country = df.groupby("Country")["CO₂ Emissions (Metric Tons per Capita)"].mean()
co2_by_country
```

```python
# Agrupar por país y calcular diferentes métricas
co2_summary = df.groupby("Country")["CO₂ Emissions (Metric Tons per Capita)"].agg(
    Total_Emissions="sum",  # Suma total de emisiones
    Mean_Emissions="mean",  # Emisiones promedio
    Record_Count="count"    # Número de registros por país
).reset_index()
co2_summary
```

```python
x=co2_summary["Country"]
print(x)
y=co2_summary["Total_Emissions"]
```

```python
import matplotlib.pyplot as plt

# Agrupar datos por páis y calcular emisiones promedio
co2_by_country = df.groupby("Country")["CO₂ Emissions (Metric Tons per Capita)"].mean()

plt.figure(figsize=(8, 5))
plt.bar(co2_by_country.index, co2_by_country.values, color="skyblue")

# Personalización
plt.title("")
plt.xlabel("Grade_Yes")
plt.ylabel("CO₂ (Toneladas Métricas per Cápita)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### ¿Qué pasa si gráfico antes de agrupar?

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo con datos sin agrupar
data = {
    "Country": ["Canada", "USA", "Germany", "India", "Brazil", "Canada", "USA", "Germany", "India", "Brazil"],
    "CO₂ Emissions (Metric Tons per Capita)": np.random.randint(1, 19, size=10)  # Generar valores aleatorios
}

df1 = pd.DataFrame(data)

# Mostrar el DataFrame
df1
```

```python
import matplotlib.pyplot as plt

# Graficar sin agrupar
plt.figure(figsize=(10, 5))
plt.bar(df1["Country"], df1["CO₂ Emissions (Metric Tons per Capita)"])

# Personalización
plt.title("Emisiones de CO₂ por País (Sin Agrupar)")
plt.xlabel("País")
plt.ylabel("CO₂ (Toneladas Métricas per Cápita)")
plt.xticks(rotation=45)
# Mostrar más barras
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```


*  El gráfico puede ser confuso, ya que hay múltiples barras para el mismo país, en lugar de una única barra representando un valor agregado.
* La información no está clara porque no sabemos si los valores representan un total, un promedio o simplemente datos individuales sin relación.
*  No hay una comparación clara entre países, ya que cada registro se muestra de manera individual en vez de consolidado.

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=co2_by_country.index, y=co2_by_country.values, palette="Blues")

# Personalización
plt.title("Emisiones Promedio de CO₂ por Continente (Seaborn)")
plt.xlabel("Continente")
plt.ylabel("CO₂ (Toneladas Métricas per Cápita)")
plt.xticks(rotation=45)
plt.show()
```

## Histrograma

Un histograma es una representación gráfica que muestra la distribución de datos en rangos. Sirve para analizar la frecuencia con la que una variable toma diferentes valores.

* Para ver la distribución de una variable numérica.
* Para detectar valores atípicos (outliers).
* Resumir grandes conjuntos de datos visualmente


Mapa de colores Matplotlib [link](https://matplotlib.org/stable/users/explain/colors/colormaps.html)

**Rice criterion**: Para escoger el número de bins o intervalos de nuestro histogramas. $k=2\sqrt[3]{n}$


```python
df.head(5)
```

```python
"""
Con Matplotlib
"""
plt.figure(figsize=(8, 5))
plt.hist(df["Sea Level Rise (mm)"], bins=10, color="orange", edgecolor="black")

plt.title("Distribución del aumento del nivel del mar")
plt.xlabel("Aumento del nivel del mar (mm)")
plt.ylabel("Frecuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
"""
Con seaborn
"""

plt.figure(figsize=(8, 5))
sns.histplot(df["Sea Level Rise (mm)"], bins="rice", kde=True, color="red")

plt.title("Distribución del aumento del nivel del mar (Seaborn)")
plt.xlabel("Aumento del nivel del mar (mm)")
plt.ylabel("Frecuencia")
plt.show()
```

 # 3. Boxplots, Scatter plots y heatmaps.

## Boxplots (Gráfico de cajas y Bigotes)

Un diagrama de cajas y bigotes es una manera conveniente de mostrar visualmente grupos de datos numéricos a través de sus cuartiles.

Las líneas que se extienden paralelas a las cajas se conocen como **bigotes,** y se usan para indicar variabilidad fuera de los cuartiles superior e inferior. Los valores atípicos se representan a veces como puntos individuales que están en línea con los bigotes. Los diagramas de cajas y bigotes se pueden dibujar vertical u horizontalmente.

```python
import seaborn as sns
plt.figure(figsize=(10, 4.5))
sns.boxplot(x="Country",y="Renewable Energy Use (%)", data=df)

#Eejrcicio agregar titulo y nombrar los ejees
```
